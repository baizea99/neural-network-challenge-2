## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
```

```python
# Create a list of at least 10 column names to use as X data
column_names = [
    'Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime',
    'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion',
    'NumCompaniesWorked'
]



# Create X_df using your selected columns
X_df = attrition_df[column_names]

# Show the data types for X_df
X_df.dtypes

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_df,   # Features
    y_df,   # Target
    test_size=0.2,  # Proportion of the data to use as the test set
    random_state=42  # Seed for reproducibility
)

X_train.head()
```

```python
# Convert your X data to numeric data types
from sklearn.preprocessing import LabelEncoder


overtime_counts = attrition_df['OverTime'].value_counts()
overtime_counts
# Create a LabelEncoder instance
label_encoder = LabelEncoder()
# Fit the LabelEncoder to the 'OverTime' column in the training data
label_encoder.fit(X_train['OverTime'])
```

```python
# Transform the 'OverTime' column in the training data
X_train['OverTime'] = label_encoder.transform(X_train['OverTime'])

# Transform the 'OverTime' column in the testing data
X_test['OverTime'] = label_encoder.transform(X_test['OverTime'])
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
onehot_encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data

y_train_department_encoded = onehot_encoder.transform(y_train[['Department']])

y_test_department_encoded = onehot_encoder.transform(y_test[['Department']])
```

```python
# Create a OneHotEncoder for the Attrition column

onehot_encoder_attrition = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data

onehot_encoder_attrition.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition_encoded = onehot_encoder_attrition.transform(y_train[['Attrition']])
y_test_attrition_encoded = onehot_encoder_attrition.transform(y_test[['Attrition']])

```

## Create, Compile, and Train the Model

```python
from tensorflow.keras.layers import Input, Dense, Concatenate

# Find the number of columns in the X training data
num_features = X_train.shape[1]

# Create the input layer
main_input = Input(shape=(num_features,), name='input')  # Adjust shape according to your main feature size


# Create at least two shared layers
shared1 = Dense(64, activation='relu', name='shared1')(main_input)
shared2 = Dense(128, activation='relu', name='shared2')(shared1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = Dense(32, activation='relu', name='department_hidden')(shared2)

# Create the output layer
department_output = Dense(3, activation='softmax', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = Dense(32, activation='relu', name='attrition_hidden')(shared2)

# Create the output layer
attrition_output = Dense(2, activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=main_input,
              outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', 
              loss={'department_output': 'mse', 'attrition_output': 'binary_crossentropy'}, 
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})
# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(
    X_train,
    {'department_output': y_train_department_encoded, 'attrition_output': y_train_attrition_encoded},
    epochs=100,  # Adjust the number of epochs as needed
    batch_size=32,  # Adjust the batch size as needed
    validation_split=0.2  # Adjust the validation split as needed
)
```

```python
# Evaluate the model with the testing data
evaluation_results = model.evaluate(
    X_test,
    {'department_output': y_test_department_encoded, 'attrition_output': y_test_attrition_encoded}
)

# Print the evaluation results to understand the structure
print(evaluation_results)
```

```python
# Print the accuracy for both department and attrition
# Unpack the evaluation results
total_loss, attrition_accuracy, department_accuracy = evaluation_results

# Print the metrics
print(f"Total Loss: {total_loss}")
print(f"Attrition Accuracy: {attrition_accuracy}")
print(f"Department Accuracy: {department_accuracy}")
```
